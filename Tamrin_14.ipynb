```python
number1 = int(input("enter first number: "))
number2 = int(input("enter second number: "))
jaam = number1 + number2
menha = number1 - number2
zarb = number1 * number2
taghsim = number1 / number2

print(number1, "+", number2, "=", jaam)
print(number1, "-", number2, "=", menha)
print(number1, "*", number2, "=", zarb)
print(number1, "/", number2, "=", taghsim)
```
