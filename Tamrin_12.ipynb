```python
class Car:

    model = ""
    color = ""
    speed = 0

    def start(self):
        print("Car started")

    def accelerate(self):
        self.speed += 10
        print("Speed:", self.speed)

    def info(self):
        print(self.model, self.color)

c = Car()

c.model = "BMW"
c.color = "White"

c.start()
c.accelerate()
c.info()
```
